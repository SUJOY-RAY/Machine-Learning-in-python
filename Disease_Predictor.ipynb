```python
from sklearn.preprocessing import LabelEncoder
disease_encoder=LabelEncoder()
symptom_encoder=LabelEncoder()
disease_dataset['Disease'] = disease_encoder.fit_transform(disease_dataset['Disease'])
disease_dataset['Symptom_1']=symptom_encoder.fit_transform(disease_dataset['Symptom_1'])
disease_dataset['Symptom_2']=symptom_encoder.fit_transform(disease_dataset['Symptom_2'])
disease_dataset['Symptom_3']=symptom_encoder.fit_transform(disease_dataset['Symptom_3'])
```

```python
X=disease_dataset.drop('Disease',axis=1)
Y=disease_dataset['Disease']
```

```python
X.head()
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.3,random_state=3,shuffle=True)
```

```python
Y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()
clf.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_pred=clf.predict(X_test)

print(accuracy_score(Y_pred,Y_test))
```
